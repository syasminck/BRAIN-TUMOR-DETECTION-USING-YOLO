```python
# Plotting the predicted image along with confidence

# Specify the folder containing the images
folder_path = '/content/drive/MyDrive/Brain_Tumor/New_Results/predict'

# Get a list of all image files in the folder
image_files = [f for f in os.listdir(folder_path) if f.endswith(('.jpg', '.png'))]

# Define the grid size (rows, columns) for displaying images
rows, cols = 4, 3

# Calculate the total number of grid positions
total_grid_positions = rows * cols

# Create a new figure for displaying the images
plt.figure(figsize=(12, 10))

# Limit the number of images to display based on grid size and maximum (12)
num_images_to_display = min(len(image_files), total_grid_positions, 12)

# Loop through each image file and display it using OpenCV and Matplotlib
for i in range(num_images_to_display):
    image_file = image_files[i]
    image_path = os.path.join(folder_path, image_file)
    image = cv2.imread(image_path)

    if image is not None:
        # Convert the image from BGR to RGB for proper display using Matplotlib
        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        # Add the image to the grid
        plt.subplot(rows, cols, i + 1)
        plt.imshow(image_rgb)
        plt.title(f"Image: {image_file}")
        plt.axis('off')
    else:
        print(f"Error reading image: {image_path}")

# Adjust spacing between subplots
plt.tight_layout()

# Show the grid of images
plt.show()
```

```python
result = results[0]
result.boxes
```

```python
for box in result.boxes:
  class_id = result.names[box.cls[0].item()]
  cords = box.xyxy[0].tolist()
  cords = [round(x) for x in cords]
  conf = round(box.conf[0].item(), 2)
  print("Object type:", class_id)
  print("Coordinates:", cords)
  print("Probability:", conf)
  print("---")
```

## **Results**

The results provided are for a YOLOv8.1.8 model trained on brain tumor data for 50 epochs. Below are the key performance metrics:

Class-wise Metrics:


**Precision** **(P): 0.956**

> A precision of 0.956 means that approximately 95.6% of the instances predicted as "all" (likely brain tumors) by the model are correct.



> A high precision indicates that when the model predicts an instance, it is accurate most of the time. In the context of brain tumor detection, this means that when the model identifies a region as a potential tumor, it is likely to be correct about 95.6% of the time.





**Recall (R): 0.833**



> A recall of 0.833 means that the model captures approximately 83.3% of the actual instances of "all" (brain tumors) present in the dataset.


> A high recall suggests that the model is effective at identifying a significant proportion of the true positive instances. In the context of brain tumor detection, this means that the model is able to find about 83.3% of the actual tumors in the images.



**mAP50 (mean Average Precision at IoU 0.5): 0.891**


> An mAP50 of 0.891 indicates a relatively high level of performance for the model at the IoU threshold of 0.5.

> The higher the mAP50 value, the better the model is at accurately localizing and classifying objects in the images.

> In practical terms, this means that, on average, the model is able to detect and localize objects with a high degree of precision and recall, considering a 50% overlap between predicted and ground truth bounding boxes.


**mAP50-95 (mean Average Precision from IoU 0.5 to 0.95): 0.559**



> A lower mAP50-95 indicates that the model might struggle more with higher IoU thresholds (closer to 0.95) where stricter criteria for bounding box overlap are applied.


> The model might be less precise in detecting objects with a high level of localization accuracy, but it could still perform well at more lenient IoU thresholds.
